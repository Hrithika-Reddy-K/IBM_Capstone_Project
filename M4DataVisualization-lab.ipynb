```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
import seaborn as sns

Query = """ 
SELECT ConvertedComp 
FROM master
"""

df = pd.read_sql_query(Query,conn)



sns.set_style("whitegrid")
plt.figure(figsize=(10,6))
plt.hist(df['ConvertedComp'], bins=30, alpha=0.7, color='skyblue')
plt.title('Distribution of Converted Compensation')
plt.xlabel('Annual Salary in USD')
plt.ylabel('Density')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
Query = """
SELECT Age
FROM master
"""

df = pd.read_sql_query(Query,conn)


plt.figure(figsize=(8, 5))
sns.boxplot(y='Age', data=df, color='pink')
plt.title('Box Plot of Age')
plt.ylabel('Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
Query = """
SELECT Age,WorkWeekHrs
FROM master
"""

df=pd.read_sql_query(Query,conn)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='WorkWeekHrs', alpha=0.7)
plt.title('Age vs. WorkWeekHrs')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
Query = """
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
"""
df=pd.read_sql_query(Query,conn)


plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='WorkWeekHrs', y='CodeRevHrs', size='Age', sizes=(20, 200), hue = 'Age', alpha=0.6, palette = 'coolwarm')
plt.title('Bubble Plot of WorkWeekHrs vs. CodeRevHrs')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
Query = """
SELECT *
FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(Query,conn)

pie_data = df['DatabaseDesireNextYear'].value_counts().head().reset_index()
pie_data.columns = ['Database', 'Count']

#for index, row in pie_data.iterrows():
    #print(f"Database: {row['Database']}, Count: {row['Count']}")
colors_list = ['pink','lightgreen','lightblue','gold','coral']
plt.figure(figsize=(6, 6))
plt.pie(pie_data['Count'], labels=pie_data['Database'], autopct='%1.1f%%', startangle=150, colors=colors_list)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()

```

```python
#Python ranking in the most desired languages to learn next year by the respondents
import sqlite3
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
import pandas as pd

Query = """
SELECT *
FROM LanguageDesireNextYear
"""

df = pd.read_sql_query(Query,conn)

top_language = df['LanguageDesireNextYear'].value_counts().reset_index()
top_language.columns = ['Language', 'Count']

# Add a column for the rank
top_language['Rank'] = top_language.index + 1


# Print the DataFrame with ranks
print(top_language)
```

```python
#respondents working currently with 'SQL'

Query = """
SELECT *
FROM LanguageWorkedWith
"""

df = pd.read_sql_query(Query,conn)
sql_count = df[df['LanguageWorkedWith'].str.contains('SQL', case=False, na=False)].shape[0]

print("Number of respondents who worked with SQL:", sql_count)
```

```python
#respondents working on 'MySQL' only
Query = """
SELECT *
FROM LanguageWorkedWith
"""

df = pd.read_sql_query(Query,conn)

# df is the DataFrame containing survey responses
# and 'LanguageWorkedWith' is the column containing information about Languages worked with

# Find respondents who work with 'MySQL' only
mysql_only_count = df[
    (df['LanguageWorkedWith'].str.contains('MySQL', case=False, na=False)) &
    (~df['LanguageWorkedWith'].str.contains('/', na=False))
].shape[0]

print("Number of respondents who work with only MySQL:", mysql_only_count)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
Query = """
SELECT WorkWeekHrs,CodeRevHrs,Age
FROM master
"""
df = pd.read_sql_query(Query,conn)

age_group_30_35 = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
stacked_data = age_group_30_35[['WorkWeekHrs', 'CodeRevHrs']].median()
stacked_data.plot(kind='bar', stacked=True, figsize=(7, 6), color = 'green')
plt.title('Median WorkWeekHrs and CodeRevHrs for Age Group 30-35')
plt.xlabel('Age Group')
plt.ylabel('Median Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
Query = """
SELECT ConvertedComp,Age
FROM master
"""
df = pd.read_sql_query(Query,conn)

median_convertedcomp = df.groupby('Age')['ConvertedComp'].median().loc[45:60]
#median_convertedcomp = df.groupby('Age')['ConvertedComp'].median().loc[25:30]
plt.figure(figsize=(8, 6))
median_convertedcomp.plot(marker='s')
plt.title('Median ConvertedComp by Age')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
Query = """
SELECT MainBranch
FROM master
"""
df = pd.read_sql_query(Query,conn)

plt.figure(figsize=(9, 6))
main_branch_counts = df['MainBranch'].value_counts()
main_branch_counts.plot(kind='barh', color='grey')
plt.title('Count of Respondents by MainBranch')
plt.xlabel('Count')
plt.ylabel('MainBranch')
plt.show()
```
